```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df1 = pd.read_csv('beam_deflection_euler_0.3m.csv')
df2 = pd.read_csv('beam_deflection_euler_3m.csv')
df3 = pd.read_csv('beam_deflection_timo_0.3m.csv')
df4 = pd.read_csv('beam_deflection_timo_3m.csv')

df5 = pd.read_csv('normal_stress_euler_0.3m.csv')
df6 = pd.read_csv('normal_stress_euler_3m.csv')
df7 = pd.read_csv('normal_stress_timo_0.3m.csv')
df8 = pd.read_csv('normal_stress_timo_3m.csv')
df9 = pd.read_csv('normal_stress_calfem_L=3m.csv')
df10 = pd.read_csv('normal_stress_calfem_L=0.3m.csv')

# Extract columns
x1 = df1['Position (m)']
w1 = df1['Deflection (mm)']

x2 = df2['Position (m)']
w2 = df2['Deflection (mm)']

x3 = df3['Position (m)']
w3 = df3['Deflection (mm)']

x4 = df4['Position (m)']
w4 = df4['Deflection (mm)']

x5 = df5['Position (m)']
sigma1 = df5['Normal Stress (MPa)']

x6 = df6['Position (m)']
sigma2 = df6['Normal Stress (MPa)']

x7 = df7['Position (m)']
sigma3 = df7['Normal Stress (MPa)']

x8 = df8['Position (m)']
sigma4 = df8['Normal Stress (MPa)']

x9 = df9['Position (m)']
sigma5 = df9['Normal Stress (MPa)']

x10 = df10['Position (m)']
sigma6 = df10['Normal Stress (MPa)']

# Plot
plt.figure(figsize=(8,5))
plt.plot(x1, w1, 'b-', linewidth=2, label='Euler')
plt.plot(x3, w3, 'r-', linewidth=2, label='Timoshenko')
plt.title('Beam Deflection L=0.3m', fontsize=14, fontweight='bold')
plt.xlabel('(m)', fontsize=12)
plt.ylabel('Deflection (mm)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.show()



# Plot
plt.figure(figsize=(8,5))
plt.plot(x2, w2, 'b-', linewidth=2, label='Euler')
plt.plot(x4, w4, 'r-', linewidth=2, label='Timoshenko')
plt.title('Beam Deflection L=3m', fontsize=14, fontweight='bold')
plt.xlabel('(m)', fontsize=12)
plt.ylabel('Deflection (mm)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.show()


plt.figure(figsize=(8,5))
plt.plot(x5, sigma1, 'r-', linewidth=2, label= 'Euler')
plt.plot(x7, sigma3, 'b-', linewidth=2, label='Timoshenko')
plt.plot(x10, sigma6, 'r--', linewidth=2, label='Calfem')
plt.title('Normal Stress σ_xx L=0.3m', fontsize=14, fontweight='bold')
plt.xlabel('(m)', fontsize=12)
plt.ylabel('Normal Stress (MPa)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.show()


plt.figure(figsize=(8,5))
plt.plot(x6, sigma2, 'r-', linewidth=2, label= 'Euler')
plt.plot(x8, sigma4, 'b-', linewidth=2, label='Timoshenko')
plt.plot(x9, sigma5, 'r--', linewidth=2, label='Calfem')
plt.title('Normal Stress σ_xx L=3m', fontsize=14, fontweight='bold')
plt.xlabel('(m)', fontsize=12)
plt.ylabel('Normal Stress (MPa)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.show()
```
