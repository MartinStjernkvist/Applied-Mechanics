```python
# Problem 1

print('a)')

x, L, xi, xj, E, A, P = sp.symbols('x, L, xi, xj, E, A, P')

Le = L / 3
k = 2 * E * A / L
b = P / L

N = 1 / L * sp.Matrix([xj - x, x - xj]).T

Be = 1 / Le  * sp.Matrix([-1, 1]).T
display(Be)

BeTBe = Be.T @ Be

Ke = E * A * BeTBe * Le
displayvar('Ke', Ke)

fle = Le / 2 * sp.Matrix([1, 1]) * b

display(fle)

K = sp.zeros(4,4)
fl = sp.zeros(4,1)
display(fl)

K[:2, :2] = Ke
K[1:3, 1:3] += Ke
K[2:, 2:] += Ke

Kc = k * N.subs(x, L).T @ N.subs(x, L) 
displayvar('Kc', Kc)
K[2:, 2:] += Kc

fl[:2, 0] = fle[:]
fl[1:3, 0] += fle
fl[2:, 0] += fle

displayvar('K', K)
displayvar('fl', fl)

bc_dofs = [1]
bc_vals = [0]

f = fl

"""
Not reasonable to have to solve this symbolically
"""

a, r = solve_eq(K, f, bc_dofs, bc_vals)
displayvar('a', a)
```

```python
# Problem 3

print('b)')

x, b, t, h0, xi, xj = sp.symbols('x, b, t, h0, xi, xj')

N1 = 1 / (b / 4) * (xj - x)
N2 = 1 / (b / 4) * (x - xj)

Ne = sp.Matrix([N1, N2])
displayvar('Ne', Ne)

hy = - h0 * (1 - x / b) 

integrand = hy * Ne * t

ans = sp.integrate(integrand, (x, xi, xj))

ans = ans.subs(xi, 0).subs(xj, b/4)
ans = ans.simplify()
displayvar('ans', ans)

"""
FEL I FACIT
"""
```
