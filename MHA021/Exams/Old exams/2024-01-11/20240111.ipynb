```python
# Problem 1

print('b)')

x, L, q0, x_i, x_j, h = sp.symbols('x, L, q0, x_i, x_j, h')

Le = x_j - x_i

N1e = (x_j - x) / Le  
N2e = (x - x_i) / Le

Ne = sp.Matrix([N1e, N2e])

qx = q0 * (1 - sp.Rational(1, 5) * (x / L))

fle = sp.integrate((Ne * qx), (x, x_i, x_j))

# fle = fle.subs(x_j, x_i + h)

fle = fle.simplify()
display(fle)
```

```python
print('c)')

# Define inputs
L_val = 1
E_val = 80e9
A0_val = 20e-4 # m^2
q0_val = 100e3 # N/m
x_val = L_val

E, A0 = sp.symbols('E, A0')

BeTBe = (1 / Le)**2 * sp.Matrix([
    [1,-1],
    [-1, 1]
])

integrand = A0 / 2 * (2 - (x / L)**2)

integral = sp.integrate(integrand, (x, x_i, x_j))
integral = integral.simplify()
displayvar('integral', integral)

Ke = E * BeTBe * integral
displayvar('Ke', Ke)

Ke_func = sp.lambdify((x_i, x_j, A0, E, L), Ke, 'numpy')
fle_func = sp.lambdify((x_i, x_j, L, q0), fle, 'numpy')

Le_val = L_val / 3

Ke1_val = Ke_func(0 * Le_val, 1 * Le_val, A0_val, E_val, L_val)
Ke2_val = Ke_func(1 * Le_val, 2 * Le_val, A0_val, E_val, L_val)
Ke3_val = Ke_func(2 * Le_val, 3 * Le_val, A0_val, E_val, L_val)

fle1_val = fle_func(0 * Le_val, 1 * Le_val, L_val, q0_val)
fle2_val = fle_func(1 * Le_val, 2 * Le_val, L_val, q0_val)
fle3_val = fle_func(2 * Le_val, 3 * Le_val, L_val, q0_val)

K = np.zeros((4, 4))
fl = np.zeros((4, 1))

K[:2, :2] = Ke1_val[:,:]
K[1:3, 1:3] += Ke2_val[:, :]
K[2:, 2:] += Ke3_val[:, :]

fl[:2] = fle1_val[:]
fl[1:3] += fle2_val[:]
fl[2:] += fle3_val[:]

bc_dofs = np.array([1])
bc_vals = np.array([0])

a, _ = solve_eq(K, fl, bc_dofs, bc_vals)
disp = a[4-1]
print(f'{disp:.2e}')

displayvar('K', K)
displayvar('f_l', fl)
displayvar('a', a)
```

```python
# Problem 3

print('b)')

# Define inputs

rho_val = 7_800
omega_val = 20
g_val = 9.81

scale = 1e-2

x4 = 3 * scale
y4 = 0 * scale
x5 = 4 * scale
y5 = 0 * scale
x17 = 3 * scale
y17 = 1 * scale
x18 = 4 * scale
y18 = 0.8 * scale

x, y, xi, eta, rho, g, omega, t = sp.symbols('x, y, xi, eta, rho, g, omega, t')

N1 = (1 - xi) * (1 - eta) / 4
N2 = (1 + xi) * (1 - eta) / 4
N3 = (1 + xi) * (1 + eta) / 4
N4 = (1 - xi) * (1 + eta) / 4

N_bar = sp.Matrix([N1, N2, N3, N4]).T
displayvar('Nbar', N_bar)

N = sp.Matrix([[N1, 0, N2, 0, N3, 0, N4, 0],
               [0, N1, 0, N2, 0, N3, 0, N4]])
displayvar('N', N)

xe = sp.Matrix([x4, x5, x18, x17])
ye = sp.Matrix([y4, y5, y18, y17])

x = N_bar @ xe
y = N_bar @ ye

b = sp.Matrix([rho * omega**2 * x, - rho * g])
displayvar('b', b)

J = sp.Matrix([x, y]).jacobian(sp.Matrix([xi, eta]))

fle = sp.integrate(sp.integrate((N.T @ b * t * sp.det(J)), (xi, -1, 1)), (eta, -1, 1))
fle = fle.subs(omega, omega_val).subs(rho, rho_val).subs(g, g_val)
fle = fle.simplify()
displayvar('f_l', fle, accuracy=3)

"""
FACIT ÄR FEL!!!!!!!!!!!
"""
```
