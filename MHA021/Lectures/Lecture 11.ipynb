## CST element without supports

```python
nodes = np.array([
    [0, 0], 
    [1, 0], 
    [0, 2]
])
D = hooke_2d_plane_stress(E=1, nu=0.3)
t = 1 # thickness

K, f = cst_element(nodes, D, t)
displayvar("K", K, accuracy=4)

M = cst_element_M(nodes, rho=1, t=t)
displayvar("M", M, accuracy=4)

omega2, phi = eigh(K, M)
displayvar("ω^2", np.round(omega2), accuracy=4)
displayvar("φ", phi, accuracy=4)
# displayvar("f", np.sqrt(np.round(omega2))/(2*np.pi), accuracy=4)
```

 Visualization of the modes

```python
# Define elements and edofs so we can use the helper function plot_deformed_mesh
elements = np.array([
    [1, 2, 3]
])
edofs = np.array([
    [1, 2, 3, 4, 5, 6]
])

mode = 1 # mode I want to display
for mode in range(0,6):
    phi_mode = phi[:, mode]
    # phi_mode = phi[:, mode-1]
    Ed = extract_dofs(phi_mode, edofs)
    fig = plot_deformed_mesh(nodes, elements, Ed, scale=.1, field="utuxotal", title=f"Mode {mode+1}", cycles=3, duration=2)
    fig.show()

```
